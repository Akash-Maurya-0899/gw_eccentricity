### Loading data

Let's now get some data in the required format, as described above, using the load_waveform function.

```python
help(load_waveform)
```

```python
waveform_keywords = {"approximant": "EccentricTD",
                     "q": 1.0,
                     "chi1": [0.0, 0.0, 0.0],
                     "chi2": [0.0, 0.0, 0.0],
                     "deltaTOverM": 0.1,
                     "Momega0": 0.01,
                     "ecc": 0.1,
                     "mean_ano": 0,
                     "phi_ref": 0,
                     "inclination": 0,
                     "include_zero_ecc": True}
```

```python
dataDict = load_waveform(**waveform_keywords)
```

We have used the keyword "include_zero_ecc" since we want to use the residual amplitude method that needs the data for the zero eccentricity counterpart

```python
plt.figure(0, figsize=(12, 4))
plt.plot(dataDict["t"], np.abs(dataDict["hlm"][(2, 2)]), label="eccentric")
plt.plot(dataDict["t_zeroecc"], np.abs(dataDict["hlm_zeroecc"][(2, 2)]), label="non eccentric")
plt.xlabel('t', fontsize=12)
plt.ylabel('$A_{22}$', fontsize=12)
plt.title('Amplitude for an eccentric waveform and its quasicircular counterpart.\n '
          'The difference gives us the residual amplitude.')
plt.legend();
```

### measuring eccentricity using all available methods

```python
# refernce times at which to measure eccentricity and mean anomaly
tref_vec = np.arange(-14000, -1000)

fig, axarr = plt.subplots(nrows=2, sharex=True, figsize=(12, 8))
for method in available_methods:
    ecc_vec, mean_ano_vec = measure_eccentricity(tref_vec, dataDict, method=method)
    if method == available_methods[0]:
        ls = 'solid'
    elif method == available_methods[1]:
        ls = 'dashed'
    else:
        ls = 'dotted'
    axarr[0].plot(tref_vec, ecc_vec, label=method, ls=ls)
    axarr[1].plot(tref_vec, mean_ano_vec, label=method, ls=ls)
axarr[1].set_xlabel("time", fontsize=12)
axarr[0].set_ylabel("eccentricity", fontsize=12)
axarr[1].set_ylabel("mean anomaly", fontsize=12)
axarr[1].grid()
axarr[0].legend(fontsize=12)
axarr[1].legend(fontsize=12)
```

### Visualizing the peaks and troughs in the data

Peaks and troughs in the frequency can be visualized using the different methods for measuring eccentricity. For example let's use the ResidualAmplitude method

```python
method = "ResidualAmplitude"
ecc_vec, mean_ano_vec, eccMethod = measure_eccentricity(tref_vec, dataDict, method="ResidualAmplitude", return_ecc_method=True)
```

```python
?eccMethod.interp_extrema
```

```python
# Get interpolator for Periastrons/Apastrons, it returns the locations used for creating the interpolator as well
omega22_peaks_interp, peaks_location = eccMethod.interp_extrema(which="peaks", spline_keywords=eccMethod.spline_keywords)
omega22_troughs_interp, troughs_location = eccMethod.interp_extrema(which="troughs", spline_keywords=eccMethod.spline_keywords)
t = eccMethod.t
omega22 = eccMethod.omega22
t_peaks = t[peaks_location]
t_troughs = t[troughs_location]
# evaluate the interpolator at the peaks and troughs
omega22_peaks = omega22_peaks_interp(t_peaks)
omega22_troughs = omega22_troughs_interp(t_troughs)

fig, ax = plt.subplots(nrows=3, figsize=(12, 12), sharex=True)

ax[0].plot(tref_vec, ecc_vec, label=method, ls="-")
ax[1].plot(tref_vec, mean_ano_vec, label=method, ls="-")
ax[1].set_xlabel("time", fontsize=12)
ax[0].set_ylabel("eccentricity", fontsize=12)
ax[1].set_ylabel("mean anomaly", fontsize=12)
ax[1].grid()
ax[0].legend(fontsize=12)
ax[1].legend(fontsize=12)

ax[2].plot(t_peaks, omega22_peaks, marker="o", label='Periastrons')
ax[2].plot(t_troughs, omega22_troughs, marker="o", label='Apastrons')
ax[2].plot(t, omega22)
ax[2].set_ylim(0, 0.06)
ax[2].set_xlim(min(tref_vec), max(tref_vec))
ax[2].set_xlabel("time", fontsize=12)
ax[2].set_ylabel("$\omega_{22}$", fontsize=12)
ax[2].grid()
ax[2].legend(fontsize=12)
ax[2].set_title('Periastron and apastrons picked by the peak finder, '
          'shown on top of the frequency.', fontsize=12);
```
